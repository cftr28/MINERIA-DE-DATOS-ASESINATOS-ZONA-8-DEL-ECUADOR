```python
import pandas as pd

# Especifica la ruta al archivo en Google Drive
file_path = '/content/drive/My Drive/ASESINATOS.xls'

# Cargar el archivo Excel en un DataFrame
df = pd.read_excel(file_path)

# Verificar las primeras filas para entender la estructura del DataFrame
print("Primeras filas del DataFrame:")
print(df.head())
```

```python
# Obtener el número de registros (filas) en el DataFrame
num_registros = df.shape[0]

print(f"El número de registros en el DataFrame es: {num_registros}")
```

```python

# Convertir las variables categóricas a variables numéricas
df['Lugar_numeric'] = df['Lugar'].astype('category').cat.codes
df['Tipo_Lugar_numeric'] = df['Tipo Lugar'].astype('category').cat.codes

# Calcular la correlación entre Lugar y Tipo Lugar
correlation = df[['Lugar_numeric', 'Tipo_Lugar_numeric']].corr()

# Mostrar la matriz de correlación
print(correlation)
import matplotlib.pyplot as plt
import seaborn as sns

# Graficar la matriz de correlación
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', cbar=True, square=True)
plt.title('Matriz de Correlación entre Lugar y Tipo de Lugar')
plt.show()

```

```python


# Convertir las variables categóricas a variables numéricas
df['Circuito_numeric'] = df['Circuito'].astype('category').cat.codes
df['Subcircuito_numeric'] = df['Subcircuito'].astype('category').cat.codes

# Calcular la correlación entre las variables
correlation = df[['Circuito_numeric', 'Subcircuito_numeric']].corr()

# Mostrar la matriz de correlación
print("Matriz de correlación entre Circuito y Subcircuito:")
print(correlation)
import matplotlib.pyplot as plt
import seaborn as sns

# Graficar la matriz de correlación
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', cbar=True, square=True)
plt.title('Matriz de Correlación entre Circuito y Subcircuito')
plt.show()
```

```python


# Convertir las variables categóricas a variables numéricas
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    df[column + '_numeric'] = df[column].astype('category').cat.codes

# Seleccionar las nuevas columnas numéricas para la correlación
numeric_columns = [col + '_numeric' for col in categorical_columns]
df_numeric = df[numeric_columns]

# Calcular la matriz de correlación
correlation_matrix = df_numeric.corr()

# Mostrar la matriz de correlación
print("Matriz de correlación:")
print(correlation_matrix)
```
